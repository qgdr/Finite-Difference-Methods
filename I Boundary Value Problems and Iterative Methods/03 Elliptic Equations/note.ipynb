```python

def plt_conv_ker(conv_ker):
    plt.imshow(conv_ker)
    plt.axis('off')
    for i, h in enumerate(conv_ker):
        for j, e in enumerate(h):
            if e != 0:
                plt.text(i, j, e, fontdict={'fontsize':24})


conv_ker1 = np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0],
])

plt_conv_ker(conv_ker1)
```

```python
conv_ker2 = np.array([
    [1, 4, 1],
    [4, -20, 4],
    [1, 4, 1]
])
plt_conv_ker(conv_ker2)
```

```python
import cv2 as cv

plt.figure(figsize=(16, 10))
loli = cv.imread("loli.png", flags=0)
print(loli.shape)

plt.subplot(131)
plt.imshow(loli, cmap='gray')
plt.axis('off')

plt.subplot(132)
conv_loli = cv.filter2D(loli, -1, conv_ker1, borderType=cv.BORDER_CONSTANT)
plt.imshow(256-conv_loli, cmap='gray')
plt.axis('off')

plt.subplot(133)
conv_loli = cv.filter2D(loli, -1, conv_ker2/6, borderType=cv.BORDER_CONSTANT)
plt.imshow(256-conv_loli, cmap='gray')
plt.axis('off')

plt.show()
```
